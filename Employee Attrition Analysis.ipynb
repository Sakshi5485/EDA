## TASK - 2 :

### Define a function that identifies the data type of columns as categorical or numerical. Separate the names of  categorical columns and numerical columns and save them in two different lists.

```python
def dtypes_col(data):
    categorical = []
    numerical = []
    for i in data.columns:
        if data[i].nunique() > 10:
            numerical.append(i)
        else:
            categorical.append(i)
            
    return categorical, numerical

categorical, numerical = dtypes_col(data)
```

```python
print(numerical)
print()
print(type(numerical))
```

```python
print(categorical)
print()
print(type(categorical))
```

#### Use the "EnvironmentSatisfaction" column dislpay the employee count by environment satisfaction score using a bar chart and display its percentage on the top of each bar.

```python
def percentage(ax, count: "number of rows in data "):
    for bar in ax.patches:                                            
        percentage = f"{round((bar.get_height() / count) *100, 2)}%"  
                                                                           
        x = bar.get_x() + bar.get_width() /2                         
        y = bar.get_height()                                         
        ax.annotate(percentage, (x, y), va = "bottom", ha = "center")  
                                                                      
                                                                       
    plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,6))
ax.set_title("Employee Count by Environment Satisfaction")
ax = sns.countplot(x=data["EnvironmentSatisfaction"])
percentage(ax,len(data))
```

### ***Interpretation : Approximately 60% of employees have an environment satisfaction scale ranging from 3 to 4, while the remaining employees fall within the 1 to 2 range.

### Perform descriptive analysis of each numerical column by defining a function to print the following information. 1) The measures of central tendency (mean, median) , 2) The standard deviation of a column, 3) The missing values of a column 4) The skewness of a column. Perform the missing value treatment, if required.

```python
def num_info(s):
    if data[s].isnull().sum() > 0:
        mean_value = data[s].mean()
        data[s].fillna(mean_value, inplace=True)

    print(f"The mean of {s} : {data[s].mean()} ")
    print(f"The median is {s} : {data[s].median()}")
    print(f"The Standard devivation of {s} : {data[s].std()}")
    print(f"Missing values in {s} are : {data[s].isnull().sum()}")
    print(f"The skewness of {s} : {data[s].skew()}")
```
