# Employee Attrition Analysis

###### 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
import matplotlib as mat
```

```python
data = pd.read_csv("DS1_C5_S2_EmployeeAttrition_Data_Practice.csv")
```

###### 

## TASK - 1 :

### Perform bivariate and multivariate analysis on categorical columns with the help of graphs/charts, as mentioned below :

#### * Analyze the business travel category-wise employee attrition using a count plot.

```python
fig, ax = plt.subplots(figsize = (10, 7))
ax.set_title("Business travel Category-wise Attrition count")
sns.countplot(x = "BusinessTravel", hue = "Attrition", data = data)
plt.show()
```

### ***Interpretation : The percentage of employees who travel rarely churned more, even though the total count of employees who travel rarely is also greater.

###### 

#### * Analyze the education field-wise employee attrition using a bar chart.

```python
fig, ax = plt.subplots(figsize = (10, 7))
ax.set_title("Education Field-wise Attrition count")
sns.countplot(x = "EducationField", hue = "Attrition", data = data)
plt.show()
```

### ***Interpretation : The percentage of employee attrition is higher among those with a life science educational background, even though the total count of employees in this field is also greater.

###### 

#### * Analyze the department-wise employee attrition using a bar chart.

```python
fig, ax = plt.subplots(figsize = (10, 7))
ax.set_title("Department-wise Attrition count")
sns.countplot(x = "Department", hue = "Attrition", data = data)
plt.show()
```

### ***Interpretation : The majority of employees who have been attrited belong to the Research & Development department.

#### * Use the "OverTime" column to display the overtime category-wise employee attrition using pie chart.

```python
emp_count = data.groupby('OverTime')['Attrition'].value_counts().unstack()

fig, ax = plt.subplots()

for i, a in enumerate(emp_count.index):
    z = emp_count.loc[a]
    ax.pie(z, labels=z.index, autopct='%0.2f%%',colors= ['lightgreen', 'lightskyblue'], explode=(0.05, 0.05))
    ax.set_title("Over Time category-wise Attrition count")

    plt.show()
```

### ***Interpretation : Those who work overtime are more attrited.

######

## TASK - 2 :

### Perform the bi-variate analysis between categorical and numerical variables using an appropriate graph/chart.

#### * Use the "Age" column to display age group-wise distribution of the employee attrition using a histogram and a box plot.

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
sns.histplot(x = data["Age"], hue = data["Attrition"], ax = ax[0])
sns.boxplot(x = data["Attrition"] ,y = data["Age"] , ax = ax[1])
plt.show()
```

### ***Interpretation : Most of the employees are between 25 to 45 years of age.

##### 

#### Use the "HourlyRate" column to display the hourly salary-wise employee attrition using a histogram and a box plot.

### ***Interpretation : Employees earning between 50 and 70 per hour are more likely to leave.
